#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
url='https://raw.githubusercontent.com/GisseYo/ChallengeTelecomX/main/TelecomX_Data.json'
bd_original=pd.read_json(url)
bd_original
```

#ðŸ”§ TransformaciÃ³n

```python
colum_customer = pd.json_normalize(bd_original['customer'])
colum_customer
```

```python
colum_phone = pd.json_normalize(bd_original['phone'])
colum_phone
```

```python
colum_internet = pd.json_normalize(bd_original['internet'])
colum_internet
```

```python
colum_account = pd.json_normalize(bd_original['account'])
colum_account
```

```python
bd_concatenar = pd.concat([bd_original, colum_customer, colum_phone, colum_internet,colum_account], axis=1)
bd_ok = bd_concatenar.drop(['customer','phone','internet','account'], axis=1)
bd_ok
```

```python
bd_ok.info()
```

```python
bd_analisis = bd_ok[['Churn','SeniorCitizen','Dependents', 'tenure', 'Contract', 'Charges.Monthly', 'Charges.Total']]
bd_analisis

```

#ðŸ“Š Carga y anÃ¡lisis

```python
bd_analisis.isnull().sum()
```

```python
bd_analisis.info()
```

```python
bd_analisis['Cuentas_Diarias']= bd_analisis['Charges.Monthly']/30
bd_analisis
```

```python
bd_analisis['Churn']= bd_analisis['Churn'].replace({'Yes': 1, 'No': 0})
bd_analisis
```

```python
bd_analisis['Dependents']= bd_analisis['Dependents'].replace({'Yes': 1, 'No': 0})
bd_analisis
```

```python
bd_analisis = bd_analisis.rename(columns={'Churn': 'Es_Cliente', 'SeniorCitizen': 'Mayor_Edad', 'Dependents': 'Tiene_Dependientes', 'tenure': 'Meses_Contrato', 'Charges.Monthly': 'Total_Mesual', 'Charges.Total': 'Total_Anual' })
bd_analisis
```

```python
bd_clientes = bd_analisis[bd_analisis['Es_Cliente'] == 1]
bd_clientes
```

```python
bd_retirados = bd_analisis[bd_analisis['Es_Cliente'] == 0]
bd_retirados
```
